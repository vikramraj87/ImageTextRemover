```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

from kv.image import Image, ImageMode
from utils import generate_subplots_for_images, load_images_from_folder
from text_detection import TextDetection


text_detector = TextDetection(num_dilations=2)
```

```python
folder = 'samples'
images = load_images_from_folder(folder)

output: [Image] = []
    
with PdfPages("{}.pdf".format(folder)) as pdf:
    for idx, image in enumerate(images):
        inpainted, mask = text_detector.detect(image)
        output += [image, inpainted, mask]
        plt.figure(figsize=(11.69, 8.27))
        _, (ax1, ax2) = plt.subplots(1, 2)
        
        ax1.set_title("Original")
        ax1.axis('off')
        ax1.imshow(image.data)
        
        ax2.set_title("Mask")
        ax2.axis('off')
        ax2.imshow(mask.data, cmap='gray')
        
        pdf.savefig(dpi=600)
        plt.close
        
        plt.figure(figsize=(11.69, 8.27))
        _, ax1 = plt.subplots(1, 1)
        ax1.set_title("Final")
        ax1.axis('off')
        ax1.imshow(inpainted.data)
        
        pdf.savefig(dpi=1000)
        plt.close()

print("PDF created")
```
